# Correlation matrix

```python
#LAbel encoding
train_data.replace({"Loan_Status":{'N':0,'Y':1},'Married':{'No':0,'Yes':1},
                    'Gender':{'Male':0,'Female':1},
                    'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2}},inplace=True)
```

```python
#plotting the correlation matrix
sns.heatmap(train_data.corr() ,cmap='cubehelix_r')
```

```python
#correlation table
corr = train_data.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

Credit_History has the highest correlation with Loan_Status (a positive correlation of 0.54). Therefore our target value is highly dependant on this column.

There is no high correlation between any of the variable
